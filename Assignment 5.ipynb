# File & Exception handling Questions


1. **What is the difference between interpreted and compiled languages?**
   - In interpreted languages, the code is executed line-by-line by an interpreter, like Python. In compiled languages, the code is translated into machine code all at once by a compiler before running, like C or Java.

2. **What is exception handling in Python?**
   - Exception handling in Python is the process of managing errors that occur during the program's execution. It lets you catch and respond to these errors gracefully without crashing the program.

3. **What is the purpose of the finally block in exception handling?**
   - The `finally` block always runs, regardless of whether an error occurred or not. It is typically used for cleanup tasks, like closing files or releasing resources.

4. **What is logging in Python?**
   - Logging in Python is a way to record events that happen during the execution of a program. It helps you track the flow of the program and debug issues.

5. **What is the significance of the `__del__` method in Python?**
   - The `__del__` method is a destructor method in Python. It is automatically called when an object is about to be destroyed to perform any necessary cleanup, like closing files or releasing resources.

6. **What is the difference between `import` and `from ... import` in Python?**
   - `import` loads the entire module, while `from ... import` allows you to import specific functions or classes from a module directly, saving memory and making the code shorter.

7. **How can you handle multiple exceptions in Python?**
   - You can handle multiple exceptions by using multiple `except` blocks or by using a single `except` block with a tuple of exceptions to catch different types of errors.

8. **What is the purpose of the `with` statement when handling files in Python?**
   - The `with` statement automatically handles opening and closing files, making sure the file is closed even if an error occurs.

9. **What is the difference between multithreading and multiprocessing?**
   - Multithreading allows multiple tasks to run in the same process, sharing memory. Multiprocessing runs tasks in separate processes, each with its own memory, making it suitable for CPU-bound tasks.

10. **What are the advantages of using logging in a program?**
    - Logging helps track program events, errors, and performance. It is essential for debugging, monitoring, and maintaining programs in production environments.

11. **What is memory management in Python?**
    - Memory management in Python is the process of allocating and deallocating memory for objects automatically. Python uses a garbage collector to manage memory usage.

12. **What are the basic steps involved in exception handling in Python?**
    - The basic steps are:
      - Try: Place code that may raise an exception.
      - Except: Handle the exception if it occurs.
      - Finally (optional): Clean up resources regardless of whether an exception occurred.

13. **Why is memory management important in Python?**
    - Memory management is important to avoid memory leaks, which can slow down or crash programs. Python’s garbage collection helps keep memory usage efficient.

14. **What is the role of `try` and `except` in exception handling?**
    - `try` runs the code that might raise an exception. `except` catches the exception and allows you to handle it (e.g., print an error message) without stopping the program.

15. **How does Python's garbage collection system work?**
    - Python automatically manages memory by tracking objects and cleaning up those that are no longer needed (garbage collection). It helps prevent memory leaks.

16. **What is the purpose of the `else` block in exception handling?**
    - The `else` block runs only if no exceptions were raised in the `try` block. It’s used for code that should run when the `try` block succeeds.

17. **What are the common logging levels in Python?**
    - The common logging levels are:
      - `DEBUG`: Detailed information, typically useful for debugging.
      - `INFO`: General information about program execution.
      - `WARNING`: Indicates something unexpected but non-fatal.
      - `ERROR`: An error occurred but the program can recover.
      - `CRITICAL`: A severe error that may cause the program to stop.

18. **What is the difference between `os.fork()` and `multiprocessing` in Python?**
    - `os.fork()` creates a new process by duplicating the current process, available only on Unix-based systems. `multiprocessing` is a cross-platform module that allows creating processes in a more flexible way.

19. **What is the importance of closing a file in Python?**
    - Closing a file ensures that all data is saved and resources are released. Failing to close files can lead to memory issues and data loss.

20. **What is the difference between `file.read()` and `file.readline()` in Python?**
    - `file.read()` reads the entire file at once, while `file.readline()` reads one line at a time from the file.

21. **What is the logging module in Python used for?**
    - The logging module is used for recording logs in your program, allowing you to track events, errors, and other useful information.

22. **What is the `os` module in Python used for in file handling?**
    - The `os` module provides functions to interact with the file system, such as creating, deleting, and checking files and directories.

23. **What are the challenges associated with memory management in Python?**
    - Challenges include managing large objects, avoiding memory leaks, and ensuring the garbage collector runs efficiently, especially with complex data structures.

24. **How do you raise an exception manually in Python?**
    - You can raise an exception using the `raise` keyword, followed by the exception type, like `raise ValueError("Invalid input")`.

25. **Why is it important to use multithreading in certain applications?**
    - Multithreading is important for running multiple tasks concurrently, which can improve the performance of I/O-bound tasks and make programs more responsive.

18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler

def setup_logging():

    log = logging.getLogger("MyLogger")
    log.setLevel(logging.DEBUG)

    handler = RotatingFileHandler(
        "app.log",
        maxBytes=1 * 1024 * 1024,  # Maximum file size: 1MB
        backupCount=3  # Keep the last 3 log files
    )

    handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))

    log.addHandler(handler)

    return log

log = setup_logging()

for i in range(2):
    log.debug(f"Debug message {i}")
    log.info(f"Info message {i}")
    log.warning(f"Warning message {i}")
    log.error(f"Error message {i}")
    log.critical(f"Critical message {i}")


## The logs will be ratated to the next file once the main file exceeds 1MB
## more details will be shown if executed in vscode or pycharm or jupyter
```

23. Write a Python program that writes to a log file when an error occurs during file handling.

```python
import logging

file = "check.txt"

logging.basicConfig(filename = "check.log", level = logging.DEBUG, format = "%(asctime)s - %(levelname)s - %(message)s")

size = os.path.getsize(file)

if size == 0:
  logging.error("File is empty: Check the log file for more info")
else:
  with open(file, "r") as f:
    content = f.read()
  logging.error("File is not empty: Check the log file for more info")

#Check the log file by running in the jupyter lab.
```
