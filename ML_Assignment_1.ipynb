### 1. **What is a parameter?**

A **parameter** is a configuration variable internal to the model that is learned from the training data. For example, in linear regression, the weights and bias are parameters.

---

### 2. **What is correlation?**

**Correlation** measures the statistical relationship between two variables, indicating how one may change with respect to the other. It ranges from -1 to +1.

---

### 3. **What does negative correlation mean?**

**Negative correlation** means that as one variable increases, the other decreases. A perfect negative correlation has a coefficient of -1.

### 17. **What is causation? Difference between correlation and causation?**

**Causation** implies one variable directly affects another.
**Correlation** shows a relationship but doesn't prove causality.

**Example**:

* Correlation: Ice cream sales and drowning increase in summer.
* Causation: Increased temperature causes both; one doesn’t cause the other.
